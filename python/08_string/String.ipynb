# String Slicing

```python
There are two main method to slice a string
1] using Build-in  **slice()**
2] using the [:] array slice

```

Syntax : 
#### slice(stop)
####  slice(start , stop ,step)

```python
String = "MahonSingh"
# using slice constructor
s1 = slice(3)
s2 = slice(1 , 6 , 2)
s3 = slice(-1, -8 , -2)

# Return Type: Returns a sliced object containing elements in the given range only. 
print(s1)
print(s2)
print(s3)

print("String slicing")
print(String[s1])
print(String[s2])
print(String[s3])
```

```python
name = 'HarshaliHood'
print(name[1:2])
print(name[1::])
print(name[:1])
print(name[::3])
print(name[3:8])
```

# String Method

######  1] str.capitalize() :- Return a copy of the string with its first character chapitalize
###### 2] str.casefold() :- Return a casefolded copy of the string.CaseFolding is similar to lowercasing but in more aggressive.
######  3] str.center(width[, fillchar]) :- Return a centered string of specified width with optional padding character.
###### 4] str.count(substring[ , start[ , end]]) :- Return the number of non overlapping occurence of a substring


```python
string = "Hello World"
print(string.capitalize())
print(string.casefold())
print(string.center(10, '*'))
```

```python
name = "abbbabbsbsbsbbsssbbb"
substring = "bs"
count = name.count(substring)
print(count)
```

######  5] str.endswith(suffix[, start[, end]]):- Returns True if the string ends with the specified suffix, otherwise False
###### 6] str.expandtabs([tabsize]):- Returns a copy of the string where tab characters are expanded using spaces.
###### 7] str.expandtabs([tabsize]):- Returns a copy of the string where tab characters are expanded using spaces.
###### 8]] str.find(substring[, start[, end]]): Returns the lowest index of the substring if found, -1 otherwise


######  9] str.isalnum(): Returns True if all characters in the string are alphanumeric.
###### 10] str.isalpha(): Returns True if all characters in the string are alphabetic.
###### 11] str.isascii(): Returns True if all characters in the string are ASCII.
###### 12] str.isdecimal(): Returns True if all characters in the string are decimal (0-9)
###### 13] str.isdigit(): Returns True if all characters in the string are digits.
###### 14] str.isidentifier(): Returns True if the string is a valid Python identifier


```python
str1 = "harshali123"
str2 = "harshali@gamil.com"

print(str1.isalnum())
print(str2.isalnum())
```

```python
str1 = 'stringMethods'
print(str1.isalpha())
print(str1.isdecimal())
print(str1.isdigit())
print(str1.isidentifier())
```
