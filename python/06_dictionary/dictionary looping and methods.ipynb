### Loop through a Dictionary

```python
student_grades = {
    'John': 85,
    'Jane': 92,
    'Bob': 78,
    'Alice': 95
}

```

```python
# print all the key names in the dictionary, one by one
for x in student_grades:
    print(x)
```

```python
#  print all the values in the dictionary , one by one

for x in student_grades:
    print(student_grades[x])
```

```python
#  loop through both keys and values by using the items() method

for x , y in student_grades.items():
    print(x, y)
```

### Copy a Dictionary
You cannot copy a dictionary simply by typing dict2 = dict1, because: dict2 will only be a reference to dict1, and changes made in dict1 will automatically also be made in dict2.

<br>There are ways to make a copy, one way is to use the built-in Dictionary method copy()

```python
student_grades = {
    'John': 85,
    'Jane': 92,
    'Bob': 78,
    'Alice': 95
}

student_info = student_grades.copy()
print(student_info)
```

```python
student_info = dict(student_grades)
print(student_info)
```
